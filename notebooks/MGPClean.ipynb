```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import calendar
%matplotlib inline
```

```python
df = pd.read_excel('../data/raw/MGP Trust 2021 FINAL.xlsx', header = 1)
df
```

```python
sales_df = df[:12].copy() #get source of $1000.68 row12-P10
sales_df.fillna(0, inplace = True) #change nan to zeros
sales_df
```

```python
months = list(calendar.month_name)[1:] #generate month list
year_month = ['2022-'+ m for m in months] #prepare to turn into datetime objects
sales_df['Months'] = year_month #set df year month 
sales_df.Months = pd.to_datetime(sales_df['Months'], format='%Y-%B') 
sales_df.set_index(sales_df.Months, drop = True, inplace = True)
sales_df.drop('Months', axis = 1, inplace = True)
```

```python
sales_df['SeaSpray'] = sales_df['SS'] + sales_df['Direct']
sales_df['WindDrift'] = sales_df['WD'] + sales_df['Direct.1']
sales_df.drop(['Direct', 'Direct.1', 'WD', 'SS'], axis = 1, inplace = True)
```

```python
sales_df.drop(['Unnamed: 0', 'Unnamed: 7'], axis = 1, inplace = True)
#sales_df.rename(columns={'Direct': 'SSDirect', 'Direct.1': 'WDDirect'}, inplace=True)
sales_df
```

```python
sales_df.to_csv('../data/interim/sales.csv')
```
