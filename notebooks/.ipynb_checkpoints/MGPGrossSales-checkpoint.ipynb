```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from py_files import fetch_data, display_data
%matplotlib inline
```

```python
df = fetch_data.get_df()
```

```python
labels = ['BellaLuna','Phoenix10', 'SeaSpray', 'WindDrift']
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
plt.title('Total Sales: ${:,.2f}' .format(np.sum(df.sum())))
y = df.sum()
ax.bar(labels, y, color = 'lightblue')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
plt.title('Sales by Month')
plt.xticks(rotation=90)
ax.plot(df)
plt.legend(labels)
plt.show()
```
